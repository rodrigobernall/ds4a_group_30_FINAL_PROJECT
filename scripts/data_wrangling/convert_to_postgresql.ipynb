```
def create_csv_tables(zipfile='', output_path='', list_nested_zips=''):
    !mkdir output_path # Create the output path
    with ZipFile(zipfile, 'r') as zipObj:
        zipObj.extractall(output_path) # Unzip the giant dataset zip file

    def process_individual_files(department, prefix='', output_path='', output_filename=''):
        file_to_transform = output_path + '/' + prefix + department + '.DTA'
        if os.path.exists(file_to_transform): # Check if the file exists
            print('I am converting this file to CSV:', file_to_transform)
            df = pd.read_stata(file_to_transform)
            os.remove(file_to_transform)  # To free up space
            df.to_csv(output_path + '/' + department + '_' + output_filename, compression='gzip')
            print('File saved as', output_path + '/' + department + '_' + output_filename)
            print(' Stata converted to CSV, Stata removed.')
        else: # If file doesn't exist, do nothing.
            pass

    for my_zip_file in list_nested_zips:
        department = my_zip_file[:2] # The DANE code of the department
        with ZipFile(output_path + '/' + my_zip_file, 'r') as zipObj:
            print('I am extracting to this path:', output_path + '/' + my_zip_file)
            zipObj.extractall(output_path) # Extract the zip file of the department
            os.remove(output_path + '/' + my_zip_file) # Remove the zip file of the department to free up space
            print(' File extracted, zip removed.')
            # Vivienda data
            process_individual_files(department, prefix='CNPV2018_1VIV_A1_', output_path=output_path, output_filename='viv_list.csv.gz')
            # Hogares data
            process_individual_files(department, prefix='CNPV2018_2HOG_A1_', output_path=output_path, output_filename='hog_list.csv.gz')
            # Fallecidos data
            process_individual_files(department, prefix='CNPV2018_3FALL_A1_', output_path=output_path, output_filename='fall_list.csv.gz')
            # Personas data
            process_individual_files(department, prefix='CNPV2018_5PER_A1_', output_path=output_path, output_filename='per_list.csv.gz')
            # Personas data
            process_individual_files(department, prefix='CNPV2018_MGN_A1_', output_path=output_path, output_filename='mgn_list.csv.gz')
    #os.remove(zipfile) # Remove the giant zip file to free up space
    
    
        

    
    
# https://gist.github.com/JamesPHoughton/0f4f269e93a2b85958d8
```
